```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.datasets import make_blobs
plt.rcParams["figure.figsize"] = (10,10)
from sklearn.linear_model import Perceptron
```

```python
X = np.array([(-1, 1), (1, 1), (2, 1), (-1, -1), (1, -1)])
y = np.array([1, 1, -1, 1, -1])
theta = np.array([-0.5, 1, -1])
xx = np.linspace(-3, 3, 1000)
yy = -(theta[1] / theta[2]) * xx - (theta[0] / theta[2])

plt.title(f'Točan perceptron model, hiperravnina zadana s {theta[0]} + {theta[1]}x_1 + {theta[2]}x_2 = 0')
plt.xlabel('x_1')
plt.ylabel('x_2')

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.plot(xx, yy)
plt.show()
```

```python
a_df = pd.read_csv('./Podaci/A.csv')
X_a = a_df[['x1','x2']].to_numpy()
y_a = a_df['y'].to_numpy()

perc= Perceptron()
perc.fit(X_a ,y_a)

print(perc.coef_[0][1])
```

```python

plt.scatter(X_a[:,0],X_a[:,1],c = y_a)
xx = np.linspace(-20,40,1000)
yy = -(perc.coef_[0][0]/perc.coef_[0][1])*xx - (perc.intercept_/perc.coef_[0][1])
plt.plot(xx,yy)
```

```python

gama = min ([abs(np.dot(x,perc.coef_[0]+perc.intercept_))/np.linalg.norm(perc.coef_) for x in X_a])
print(gama)
```

```python
nls_df = pd.read_csv('./Podaci/nonlinsep.csv')
X_nls = nls_df[['x1', 'x2']].to_numpy()
y_nls = nls_df['y'].to_numpy()
plt.scatter(X_nls[:,0],X_nls[:,1],c=y_nls)
```

```python
m = X_nls.shape[0]
X_nls_tr = np.empty((m, 3))
for i in range(m):
    X_nls_tr[i, :2] = X_nls[i]
    X_nls_tr[i, 2] = np.sum(np.square(X_nls[i]))
p_nls = Perceptron()
p_nls.fit(X_nls_tr, y_nls)
p_nls.score(X_nls_tr, y_nls)
p_nls.coef_
```
