## Zadatak 1 (5 bodova)
Zadana je hiperravnina $\Theta{}^Tx + \theta_0 = 0$ koja ne prolazi kroz ishodište. Dokažite da je vektor $\Theta$ okomit na tu hiperravninu.

#### Kazemo da je vektor okomit na ravninu ako je okomit za svaki ne-nul vektor te ravnine 
#### Uzmimo dva proizvoljna vekora a i b za koje vrijedi 
$\theta^T a + \theta_0 = 0 \  i \  \Theta^T b + \theta_0 = 0$ gdje je $ a \neq b$
#### Ako je vektor okomit na a i b ne-nul vekore mora biti i na njihovu linearnu kombinaciju
#### Pogledajmo (a - b)
---
Uzmimo dvije proizvoljne točke hiperravnine $a,b$
$$
 \begin{equation}
  \left\{
    \begin{array}{rcr}
      \Theta^T a & = & -\theta_0 \\
      \Theta^T b & = & -\theta_0 \\
    \end{array}
  \right.
  , \quad b  \ne  a
  \end{equation}
$$
Kako smo uzeli da su a i b pripadaju hiperravnini onda pripada i njihova linearna kombinacija 
Možemo definirati vektor $(b-a)$.

Za neki vektor kažemo da je okomit na hiperravninu ako je okomit na bilo koji ne-nul vektor te hiperravnine.

Vektor $\Theta$ je okomit na hiperravninu ako je okomit na proizvolji vektor $(b-a)$, tj. vrijedi
$$
  \begin{align*}
    \Theta^T (b-a) & = 0 \\
    \Theta^T b - \Theta^T a & = 0 \\
    -\theta_0 - (-\theta_0) & = 0, \\
    0 = 0
  \end{align*}
$$

## Zadatak 2 (5 bodova)
Dokažite da je udaljenost točke $y \in \mathbb{R}^n$ do hiperravnine zadane s $\Theta^Tx = 0$ jednaka
$$
\left|\frac{\Theta^Ty}{||\Theta||}\right|.
$$

---

Svaku točku $y$ u nekom hiperprostoru kojeg separira hiperravnina možemo parametrizirati:
$$
    y(0) = y \\
    \Theta^Ty(\xi) = 0, \xi \in \R
$$
Gdje $\xi$ zapravo predstavlja udaljenost između točaka $y(0)$ i $y(\xi)$.

$y(\xi)$ se definira kao:
$$
    y(\xi) = 
    \left\{
      \begin{array}{rcr}
        y(0) - \xi \frac{\Theta}{||\Theta||}, & \textrm{y(0) se nalazi u pozitinom dijelu hiperravnine}\\
        y(0) + \xi \frac{\Theta}{||\Theta||}, & \textrm{inače}
      \end{array}
    \right.
$$

Kako $y(\xi)$ leži na hiperravnini vrijedi $\Theta^T y(\xi) = 0$
$$
  \begin{align*}
  0 &= \Theta^T y(\xi) &\\
    &= \Theta^T (y(0) \pm \xi \frac{\Theta}{||\Theta||}) &\\
    &= \Theta^T y(0) \pm \xi \frac{\Theta^T  \Theta }{||\Theta||} &\\
    &= \Theta^T y \pm \xi \frac{||\Theta||^2}{||\Theta||} &\\
    &= \Theta^T y \pm \xi ||\Theta|| 
  \end{align*}
$$
$$
\begin{align*}
  \Rightarrow \quad 
    \xi ||\Theta|| &= \pm \Theta^T y \\
    \xi &= \left|\frac{\Theta^Ty}{||\Theta||}\right|
  \end{align*}

$$

 2. Implementirajte funkciju *accuracy*, koja za dane podatke $X$ i klase $y$ računa udio točaka koje je perceptron točno klasificirao. Preporučam da funkcije iz ovog i prethodnog podzadatka enkapsulirate u klasu.
 3. Učitajte skupove podataka *A.csv*, *B.csv* i *C.csv*. U prva dva stupca se nalaze koordinate točaka $x$, dok su u zadnjem stupcu oznake klasa. Na svakom od skupova pokrenite perceptron algoritam i ispišite dobivene vrijednosti $\Theta$ i $k$. Uvjerite se da su podaci točno (ili netočno) klasificirani koristeći funkciju iz podzadatka 2.
 4. Grafički prikažite sva tri skupa podataka i pripadne hiperravnine $\Theta{}^Tx = 0$. Podaci iz različitih klasa trebaju imati različite boje. 
 5. Izračunajte vrijednosti geometrijskih margina $\gamma_{\textrm{geom}}$ za sva tri skupa podataka. Argumentirajte zašto je teže klasificirati točke iz skupa B nego točke iz skupa A.

```python
def fit(X, y, max_iter=5000):
    theta = np.zeros(X.shape[1])
    m = X.shape[0]
    k = 0
    while True:
        flag = True
        for i in range(m):
            if(k > max_iter):
                return theta, -1

            if y[i] != np.sign(theta.T @ X[i]):
                theta = theta + y[i] * X[i]
                k += 1
                flag = False

        if flag: 
            return theta, k
def accuracy(theta, X, y):
    count = 0
    m = X.shape[0]
    for i in range(m):
        if y[i] != np.sign(theta.T @ X[i]):
            count+=1
    return 1 - (count / m)

def plot_decision_boundary(X, theta):
    
    # X --> Inputs
    # theta --> parameters
    
    # The Line is y=mx+c
    # So, Equate mx+c = theta0.X0 + theta1.X1 + theta2.X2
    # Solving we find m and c
    x1 = [min(X[:,0]), max(X[:,0])]
    m = -theta[1]/theta[2]
    c = -theta[0]/theta[2]
    x2 = m*x1 + c
    
    # Plotting
    fig = plt.figure(figsize=(10,8))
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "r^")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs")
    plt.xlabel("feature 1")
    plt.ylabel("feature 2")
    plt.title('Perceptron Algorithm')
    plt.plot(x1, x2, 'y-')
```
